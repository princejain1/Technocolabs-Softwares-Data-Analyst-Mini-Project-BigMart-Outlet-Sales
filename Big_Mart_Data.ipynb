```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading data
df=pd.read_csv('/content/Train.csv')
df.head(5)
```

```python
# Cleaning data
df.isnull().sum()
```

```python
df[df['Item_Weight'].isnull()]
```

```python
df=df.fillna(method='bfill')
df.isnull().sum()
```

```python
df.head(10)
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('low fat','Low Fat')
```

```python
df['Item_Fat_Content'].unique()
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('LF','Low Fat')
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace('reg','Regular')
```

```python
df['Item_Fat_Content'].unique()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(df, x='Item_Type', y='Item_Outlet_Sales')

plt.title('Item Type Sales')
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df['Outlet_Establishment_Year'], y= df['Item_Outlet_Sales'])
```

```python
fat_content_dummies= pd.get_dummies(df[['Item_Fat_Content']],drop_first=True)
fat_content_dummies
```

```python
item_type_dummies= pd.get_dummies(df[['Item_Type']],drop_first=False)
item_type_dummies
```

```python
outlet_type_dummies= pd.get_dummies(df[['Outlet_Type']],drop_first=False)
outlet_type_dummies
```

```python
outlet_location_dummies= pd.get_dummies(df[['Outlet_Location_Type']],drop_first=False)
outlet_location_dummies
```

```python
df1=pd.concat([df,fat_content_dummies,item_type_dummies,outlet_type_dummies,outlet_location_dummies], axis=1)
df1
```
